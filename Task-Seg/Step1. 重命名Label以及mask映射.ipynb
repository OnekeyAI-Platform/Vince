```python
import os
from PIL import Image
import numpy as np

root = r'C:\Users\onekey\Desktop\new\imag\Mask'
fdir = r'C:\Users\onekey\Desktop\new\imag\masks'
os.makedirs(fdir, exist_ok=True)
label_set = set()
mapping = {76: 1, 150: 2, 174: 3, 226: 4}


def get_color_map_list(num_classes):
    """ Returns the color map for visualizing the segmentation mask,
        which can support arbitrary number of classes.
    Args:
        num_classes: Number of classes
    Returns:
        The color map
    """
    color_map = num_classes * [0, 0, 0]
    for i in range(0, num_classes):
        j = 0
        lab = i
        while lab:
            color_map[i * 3] |= (((lab >> 0) & 1) << (7 - j))
            color_map[i * 3 + 1] |= (((lab >> 1) & 1) << (7 - j))
            color_map[i * 3 + 2] |= (((lab >> 2) & 1) << (7 - j))
            j += 1
            lab >>= 3
    color_map = [color_map[i:i + 3] for i in range(0, len(color_map), 3)]
    return color_map

for sample in os.listdir(root):
    fname = sample.split('_')[0]
    img = Image.open(os.path.join(root, sample))
    img_arr = np.array(img)
    label_set |= set(np.unique(img_arr))
#     if 1 in set(np.unique(img_arr)):
#         print(np.sum(img_arr == 1))
    out_arr = np.zeros_like(img_arr)
    for m, k in mapping.items():
        out_arr[img_arr == m] = k
    image_slice = Image.fromarray(out_arr).convert('L')
    colors = np.array(get_color_map_list(5)).reshape(-1).tolist()
    image_slice.putpalette(colors)
    image_slice.save(os.path.join(fdir, f"{fname}"))
    
```

# 随机划分

```python
import os
from PIL import Image
import numpy as np
import shutil
import random

root = r'D:/20230428-Vince/train/'
img_dir = os.path.join(root, 'image')
msk_dir = os.path.join(root, 'mask')

to_img_dir = img_dir.replace('train', 'val')
to_msk_dir = msk_dir.replace('train', 'val')
os.makedirs(to_img_dir, exist_ok=True)
os.makedirs(to_msk_dir, exist_ok=True)

samples = os.listdir(img_dir)
random.seed(0)
random.shuffle(samples)
test_ratio = 0.2
for sample in samples[:int(test_ratio * len(samples))]:
    shutil.move(os.path.join(img_dir, sample), to_img_dir)
    shutil.move(os.path.join(msk_dir, sample), to_msk_dir)
```
